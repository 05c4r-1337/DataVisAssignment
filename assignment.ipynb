````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.basemap import Basemap

df_verify = pd.read_pickle('data/verify.pkl')
pd.reset_option('display.max_rows')      
pd.set_option('display.max_columns', None)

df_athletes = pd.read_csv('data/athlete_events.csv')
df_geo = pd.read_csv('data/geo.csv')
df_noc_iso3 = pd.read_csv('data/noc_iso3.csv')
df_noc_regions = pd.read_csv('data/noc_regions.csv')


dfs = {'df_athletes': df_athletes,
       'df_geo': df_geo,
       'df_noc_iso3': df_noc_iso3,
       'df_noc_regions': df_noc_regions}


for name, df in dfs.items():
    print(f"\n=== {name} ===")
    print("Columns:", df.columns.tolist())
    print("Number of rows:", len(df))
    print("Data types:\n", df.dtypes)
    print("Missing values:\n", df.isna().sum())
    print("Sample data (first 100 rows):\n", df.head(100))
````

````python
df_geo['ISO3'] = df_geo['Alpha-3 code'].astype(str).str.replace('"','').str.strip().str.upper()
df_geo['Country'] = df_geo['Country'].astype(str).str.replace('"','').str.strip()
df_geo['Latitude'] = df_geo['Latitude (average)'].astype(str).str.replace('"','').str.strip().astype(float)
df_geo['Longitude'] = df_geo['Longitude (average)'].astype(str).str.replace('"','').str.strip().astype(float)
df_geo = df_geo.drop_duplicates(subset='ISO3')
df_geo = df_geo[['ISO3','Country','Latitude','Longitude']]

df = df_athletes.merge(df_noc_iso3, how='left', on='NOC')

df = df.merge(df_noc_regions[['NOC','region']], how='left', on='NOC')
df.rename(columns={'region':'Region'}, inplace=True)

df = df.merge(df_geo, how='left', on='ISO3')

df['BMI'] = df['Weight'] / ((df['Height']/100) ** 2)

columns_order = ['ID','Name','Sex','Age','Height','Weight','BMI','Team','NOC',
                 'Region','ISO3','Country','Latitude','Longitude','Games','Year',
                 'Season','City','Sport','Event','Medal']

df = df[columns_order]

df = df.drop_duplicates()

print("Total rows:", df.shape[0])
print("Missing Country:", df['Country'].isna().sum())
print("Missing Latitude:", df['Latitude'].isna().sum())
print("Missing Longitude:", df['Longitude'].isna().sum())
print("Missing BMI:", df['BMI'].isna().sum())
````

## Task 2.5 Outliers

How many outliers are there in the `Age` feature (column)?
- Numerous methods exist to detect [Outliers](https://en.wikipedia.org/wiki/Outlier) mathematically.
- One such method uses the [1.5 x IQR Rule](https://www.khanacademy.org/math/statistics-probability/summarizing-quantitative-data/box-whisker-plots/a/identifying-outliers-iqr-rule), where **IQR** is the **Inter Quartile Range**.
  - The first Quartile (`Q1`) is the same as the `25th Percentile`.
  - The second Quartile (`Q2`) is the same as the `50th Percentile`.
  - The third Quartile (`Q3`) is the same as the `75th Percentile`.
  - The fourth Quartile (`Q4`) is the same as the `100th Percentile`.
  - The Inter Quartile Range (`IQR`) is defined as: `IQR = Q3 - Q1`.
  - The `1.5 x IQR Rule` states that:
    - Any data point that lies below `Q1 - 1.5 * IQR` is considered an outlier.
    - Any data point that lies above `Q3 + 1.5 * IQR` is considered an outlier.

Pandas `Series` have a method `quantile()` to calculate Quantiles.
- `Q1 = quantile(0.25)`
- `Q3 = quantile(0.75)`

Note on **Quantiles** vs **Quartiles** vs **Percentiles**
- A **Quantile** is a general partition (subdivision) of values, e.g. a 0.05, 0.10, 0.25, or 0.76 fraction of the values.
- A **Quartile** is also a type of **Quantile**, but **Quartiles** partition values into four disjoint subsets 1st, 2nd, 3rd and 4th quarter of the values.
- A **Percentile** is also a type of **Quantile**, but **Percentiles** partition values between 0% to 100% of the values.

## Task 3.5 Visualize the Distribution in Height of Male and Female Athletes

<details>
<summary>View a similar (not necessarily identical) visualization so you know what you are aiming for (click to expand)</summary>

> <img src="figures/task_3_5.png" width=600></img>

</details>

````python
import matplotlib.pyplot as plt
import numpy as np

# Heights
women_heights = df.loc[df["Sex"] == "F", "Height"]
male_heights = df.loc[df["Sex"] == "M", "Height"]

# Bins
bins = np.linspace(min(df["Height"].dropna()), max(df["Height"].dropna()), 30)
bin_centers = (bins[:-1] + bins[1:])/2 

# Histogram counts
women_counts, _ = np.histogram(women_heights, bins=bins)
male_counts, _ = np.histogram(male_heights, bins=bins)

# Convert to percentages
women_freq = women_counts / women_counts.sum()
male_freq = male_counts / male_counts.sum()

# Plot
plt.plot(bin_centers, women_freq, label="Women", color="pink")
plt.plot(bin_centers, male_freq, label="Men", color="skyblue")
plt.title("Height Distribution of Athletes")
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.grid(alpha=0.7)
plt.legend()
plt.show()
````
